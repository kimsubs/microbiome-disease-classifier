```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path

%matplotlib inline
```

## Load Data

```python
raw_path = './data/abundance_stoolsubset.txt'
raw_df = pd.read_csv(raw_path, sep='\t', dtype=str)
```

```python
# inspect df shape and structure
print(raw_df.shape)
raw_df.head(5)
```

## Clean Header

```python
# transpose df to have rows as dataset_name
df = raw_df.T

# and set the first row as the new headers
df.columns = df.iloc[0]
df = df[1:]

# inspect updated df
df.head(5)
```

## Split Metadata vs Taxa

```python
# metadata columns to keep
metadata_cols = ['sampleID', 'subjectID', 'bodysite', 'disease', 'age', 'gender', 'country', 'bmi_class']
metadata_df = df[metadata_cols].copy()
metadata_df
```

```python
# taxa columns
taxa_cols = [col for col in df.columns if col.startswith('k_')]
taxa_df = df[taxa_cols].copy()
taxa_df
```
